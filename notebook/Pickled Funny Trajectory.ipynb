These are the data collected from running the robot with a funny trajectory. We will read them from disk:

````python
import pickle
import numpy as np
traj = pickle.load(open('funny_traj.pkl', 'rb'))
````

````python
%matplotlib inline
import matplotlib.pyplot as plt
````

The data are in a numpy array:

````python
traj
````

````python
traj.shape
````

Each row contain these data, for N timesteps

```
0. Time
1. X position
2. Y Position
3. Euler Rot x
4. Euler Rot y
5. Euler rot z
6. Control signal left motor
7. Control signal right motor
```

````python
plt.plot(traj.T[0], traj.T[6])
plt.plot(traj.T[0], traj.T[7])
plt.title('Control signals of Funny Trajectory')
plt.xlabel('Time(s)')
plt.ylabel('Motor Throttle')
````

````python
plt.scatter(traj.T[1],
            traj.T[2],
            linewidths=0,
            c=traj.T[0])
plt.colorbar() # Color is time.
plt.title('Funny Trajectory Path in 2D')
plt.xlabel('x(m)')
plt.ylabel('y(m)')
````

````python
plt.scatter(traj.T[1],
            traj.T[2],
            linewidths=0,
            c=traj.T[5])
plt.colorbar() # Color is rotation.
plt.title('Funny Trajectory Orientation')
plt.xlabel('x(m)')
plt.ylabel('y(m)')
````
